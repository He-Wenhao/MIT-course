```python
import matplotlib.pyplot as plt

def f_prime(x):
    return 3*x**2 - 24*x + 10

def f_double_prime(x):
    return 6*x - 24

def newton_update(x):
    return x - f_prime(x)/f_double_prime(x)

# Iterating from starting point x0=10
x0 = 10
x_values_10 = [x0]
for i in range(30):
    x0 = newton_update(x0)
    x_values_10.append(x0)

# Iterating from starting point x0=2.5
x0 = 2.5
x_values_25 = [x0]
for i in range(30):
    x0 = newton_update(x0)
    x_values_25.append(x0)

plt.figure(figsize=(10, 6))
plt.plot(x_values_10, label='Starting at 10', marker='o')
plt.plot(x_values_25, label='Starting at 2.5', marker='x')
plt.title("Convergence of Newton's Method for f(x)")
plt.xlabel('Iteration')
plt.ylabel('x Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def g_prime(x):
    return 3*x**2 + 2

def g_double_prime(x):
    return 6*x

def newton_update_g(x):
    return x - g_prime(x)/g_double_prime(x)

# Iterating from starting point x0=5
x0 = 5
x_values_5 = [x0]
for i in range(30):
    x0 = newton_update_g(x0)
    x_values_5.append(x0)

plt.figure(figsize=(10, 6))
plt.plot(x_values_5, label='Starting at 5', marker='o')
plt.title("Convergence of Newton's Method for g(x)")
plt.xlabel('Iteration')
plt.ylabel('x Value')
plt.legend()
plt.grid(True)
plt.show()
```
