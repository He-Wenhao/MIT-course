```python
import numpy as np
import matplotlib.pyplot as plt

# Defining the functions
def f(x):
    return x**3 - 12*x**2 + 10*x + 3

def g(x):
    return x**3 + 2*x

# Generating x values
x = np.linspace(-5, 15, 400)

# Calculating y values for each function
y_f = f(x)
y_g = g(x)

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(x, y_f, label='f(x) = $x^3 - 12x^2 + 10x + 3$', color='blue')
plt.plot(x, y_g, label='g(x) = $x^3 + 2x$', color='red')

plt.title("Plot of f(x) and g(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
