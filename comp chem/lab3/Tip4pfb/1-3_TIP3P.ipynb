```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
R_lst = []
O_lst = []
File = open("water_rdf.csv")
Reader = csv.reader(File)
Reader = list(Reader)
for i in Reader[1:]:
    R_lst.append(float(i[0]))
    O_lst.append(float(i[1]))
plt.plot(R_lst,O_lst)
plt.title('O-O radial distribution function')
plt.xlabel('r($\AA$)')
plt.ylabel('g(r)')
plt.show()
```

```python
# find the peak
peak_val = max(O_lst)
peak_ind = O_lst.index(peak_val)

# find the first minumun
min_val = min(O_lst[peak_ind:])
min_ind = O_lst.index(min_val)

print('min_Distance',R_lst[min_ind])
```
