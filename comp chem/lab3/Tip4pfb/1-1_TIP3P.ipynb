```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
t_lst = np.linspace(0,5,5000)
```

```python
x_lst = []
name = 'DENSITY'
File = open("summary."+name+".csv")
Reader = csv.reader(File)
Reader = list(Reader)
for i in Reader[1:]:
    x_lst.append(float(i[0]))
print(len(x_lst))
print('AVERAGE '+name+':',sum(x_lst)/len(x_lst))
plt.plot(t_lst,x_lst)
plt.title('TIP4pfb: INSTANTANEOUS '+name)
plt.xlabel('TIME($ns$)')
plt.ylabel(name+'($g cm^{-3}$)')
plt.show()
```
