```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
t_lst = np.linspace(0,5000,2500)
```

```python
x_lst = []
File = open("diffusion.csv")
Reader = csv.reader(File)
Reader = list(Reader)
for i in Reader[1:]:
    x_lst.append(float(i[1]))
print(len(x_lst))
t_lst = np.array(t_lst).reshape((len(t_lst),1))

# do regression
model = LinearRegression()
model.fit(t_lst,x_lst)
x_pred = model.predict(t_lst)
slope = model.coef_[0]
print('D coefficient:',slope*10/6)
# do plot
plt.scatter(t_lst,x_lst,s=2,label = 'data points')
plt.plot(t_lst,x_pred,color = 'red',label = 'linear regression')
plt.title('TIP4pfb:displacement square - time')
plt.xlabel('TIME($ps$)')
plt.ylabel('$\Delta R^2 (\AA^2)$')
plt.legend()
plt.show()
```
