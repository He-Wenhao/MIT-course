```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
frame_lst = []
UV_lst = []
File = open("hb.dat")
Reader = csv.reader(File)
Reader = list(Reader)
for i in Reader[1:]:
    i_lst = i[0].strip().split()
    frame_lst.append(float(i_lst[0]))
    UV_lst.append(float(i_lst[2]))
    
print('average:',sum(UV_lst)/len(UV_lst))
plt.plot(frame_lst,UV_lst)
plt.title('TIP3p:hydrogen bonds')
plt.xlabel('frame')
plt.ylabel('#bound')
plt.show()
```
