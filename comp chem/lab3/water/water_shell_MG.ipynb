```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
F_lst = []
Shell1_lst = []
Shell2_lst = []
File = open("waterShell_MG_new")
Reader = csv.reader(File)
Reader = list(Reader)
for i in Reader[1:]:
    i_lst = i[0].strip().split()
    F_lst.append(float(i_lst[0]))
    Shell1_lst.append(float(i_lst[1]))
    Shell2_lst.append(float(i_lst[2]))
plt.plot(F_lst,Shell1_lst,label = 'shell 1')
plt.plot(F_lst,Shell2_lst,label = 'shell 2')
plt.title('O-MG radial distribution function')
plt.xlabel('Frame')
plt.ylabel('# water molecule')
plt.legend()
plt.show()

print(sum(Shell1_lst)/len(Shell1_lst))
print(sum(Shell2_lst)/len(Shell2_lst))
```
