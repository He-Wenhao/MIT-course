```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
R_lst = []
O_lst = []
File = open("water_rdf_OMG")
Reader = csv.reader(File)
Reader = list(Reader)
for i in Reader[1:]:
    i_lst = i[0].strip().split()
    R_lst.append(float(i_lst[0]))
    O_lst.append(float(i_lst[1]))
plt.plot(R_lst,O_lst)
plt.title('O-MG- radial distribution function')
plt.xlabel('r($\AA$)')
plt.ylabel('g(r)')
plt.show()
```

```python
# find the peak
peak_val = max(O_lst)
peak_ind = O_lst.index(peak_val)

# find the first minumun
min_val = min(O_lst[peak_ind:])
min_ind = O_lst[peak_ind:].index(min_val)+peak_ind
print('peak_Distance1',R_lst[peak_ind])
print('min_Distance1',R_lst[min_ind])

# find the peak
peak_val = max(O_lst)
peak_ind = O_lst.index(peak_val)

# find the first minumun
min_val = min(O_lst[peak_ind:])
min_ind = O_lst[peak_ind:].index(min_val)+peak_ind
print('peak_Distance1',R_lst[peak_ind])
print('min_Distance1',R_lst[min_ind])
```
